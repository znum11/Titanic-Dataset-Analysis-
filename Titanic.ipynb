```python
DF.dropna(subset=['embark_town'], inplace=True)
```

```python
le=LabelEncoder()
DF['survived'] = DF['survived'].map({0: 'No', 1: 'Yes'})
```

```python
DF.isnull().any()
```

```python
DF
```

```python
DF.dtypes
```

```python
sns.histplot(data=DF, x='age', hue='survived', kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
sns.countplot(data=DF, x='class', hue='survived')
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
sns.scatterplot(data=DF, x='age', y='fare', hue='survived', alpha=0.6)
plt.title('Fare Paid vs. Age (Survival Status)')
plt.show()
```

```python
sns.kdeplot(data=DF, x='fare', hue='class', fill=True, alpha=0.6, palette='viridis')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
sns.scatterplot(data=DF, x='age', palette="crest", y='fare', hue='survived', style='class', alpha=0.7)
plt.title('Age vs. Fare (Colored by Survival, Styled by Class)')
plt.show()
```

```python
sns.lineplot(data=DF.groupby('age')['fare'].mean().reset_index(), x='age', y='fare')
plt.title('Average Fare by Age')
plt.show()
```

```python
sns.barplot(data=DF, x='class', y='age', hue='survived', errorbar=None, palette='muted')
plt.title('Average Age by Class and Survival')
plt.show()
```

```python
sns.countplot(data=DF, x='embark_town', hue='survived', palette='Set2')
plt.title('Passenger Count by Embarked Town')
plt.xlabel("Embarked Town")
plt.ylabel("No. Of Passenger")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=DF, x='survived', y='age', hue='sex', palette='rocket')
plt.title('Age Distribution by Survival and Gender')
plt.show()
```

```python
sns.violinplot(data=DF, x='class', y='fare', hue='survived', split=True, palette='crest')
plt.title('Fare Distribution by Class and Survival')
plt.show()
```

```python
sns.regplot(data=DF, x='age', y='fare', scatter_kws={'alpha':0.3}, line_kws={'color':'yellow'})
plt.title('Regression: Age vs. Fare')
plt.show()
```

```python
sns.pairplot(data=DF[['age', 'fare', 'pclass', 'survived']], hue='survived', corner=True)
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

```python
sns.jointplot(data=DF, x='age', y='fare', kind='hex', cmap='Blues')
plt.suptitle('Age vs. Fare with Distributions', y=1.05)
plt.show()
```

```python
sns.catplot(data=DF, x='class', y='fare', hue='survived', col='sex', kind='box', height=4, aspect=0.7)
plt.suptitle('Fare Distribution by Class, Survival, and Gender', y=1.05)
plt.show()
```

```python
sns.stripplot(data=DF, x="class", y="age", hue="survived", palette="crest")
plt.title("Age by Class and Survival Status")
plt.legend(title="Survived")
plt.show()
```
